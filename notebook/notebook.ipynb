```python
import pandas as pd
import sqlite3 
```

```python
df = pd.read_csv('../dataset/EAFC26.csv')
```

```python
conn = sqlite3.connect(':memory:')
```

```python
df.to_sql('eafc26', conn, index=False)
```

```python
# Tabel A: Profil Pemain
df[['ID', 'Name', 'GENDER', 'Age', 'Nation', 'League', 'Team', 'Position']].to_sql('players', conn, index=False)
# Tabel B: Statistik Pemain
df[['ID', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].to_sql('stats', conn, index=False)

print("Database siap digunakan")
```

```python
query_dasar = """
SELECT 
    Name, 
    Age, 
    Team, 
    OVR, 
    Finishing 
FROM 
    eafc26
WHERE 
    Age <= 23 AND Finishing > 85
ORDER BY 
    Finishing DESC
LIMIT 5;
"""
# Menjalankan query dan menampilkannya dalam format rapi
pd.read_sql(query_dasar, conn)
```

```python
query_agregasi = """
SELECT
    Nation,
    COUNT(ID) AS Total_Players,
    AVG(OVR) AS Average_Rating
FROM
    eafc26
GROUP BY
    Nation
HAVING
    COUNT(ID) < 30
ORDER BY
    Average_Rating DESC;
"""
pd.read_sql(query_agregasi, conn)
```
