```python
import pandas as pd
import sqlite3 
```

```python
df = pd.read_csv('../dataset/EAFC26.csv')
```

```python
conn = sqlite3.connect(':memory:')
```

```python
df.to_sql('eafc26', conn, index=False)
```

```python
# Tabel A: Profil Pemain
df[['ID', 'Name', 'GENDER', 'Age', 'Nation', 'League', 'Team', 'Position']].to_sql('players', conn, index=False)
# Tabel B: Statistik Pemain
df[['ID', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].to_sql('stats', conn, index=False)

print("Database siap digunakan")
```
