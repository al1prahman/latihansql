```python
import pandas as pd
import sqlite3 
```

```python
df = pd.read_csv('../dataset/EAFC26.csv')
```

```python
conn = sqlite3.connect(':memory:')
```

```python
df.to_sql('eafc26', conn, index=False)
```

```python
# Tabel A: Profil Pemain
df[['ID', 'Name', 'GENDER', 'Age', 'Nation', 'League', 'Team', 'Position']].to_sql('players', conn, index=False)
# Tabel B: Statistik Pemain
df[['ID', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].to_sql('stats', conn, index=False)

print("Database siap digunakan")
```

```python
query_dasar = """
SELECT 
    Name, 
    Age, 
    Team, 
    OVR, 
    Finishing 
FROM 
    eafc26
WHERE 
    Age <= 23 AND Finishing > 85
ORDER BY 
    Finishing DESC
LIMIT 5;
"""
# Menjalankan query dan menampilkannya dalam format rapi
pd.read_sql(query_dasar, conn)
```

```python
query_agregasi = """
SELECT
    Nation,
    COUNT(ID) AS Total_Players,
    AVG(OVR) AS Average_Rating
FROM
    eafc26
GROUP BY
    Nation
HAVING
    COUNT(ID) < 30
ORDER BY
    Average_Rating DESC;
"""
pd.read_sql(query_agregasi, conn)
```

```python
query_join = """
SELECT 
    p.Name, 
    p.Position, 
    p.Team, 
    s.OVR, 
    s.PAC
FROM 
    players p
INNER JOIN 
    stats s ON p.ID = s.ID
WHERE 
    s.PAC > 85 AND p.GENDER = 'men'
ORDER BY 
    s.PAC DESC
LIMIT 5;
"""
pd.read_sql(query_join, conn)
```

```python
query_tim = """
SELECT
    Name,
    Team,
    OVR,
    Nation
FROM
    eafc26
WHERE
    Team = 'Real Madrid'
"""
pd.read_sql(query_tim, conn)
```

```python
query_liga = """
SELECT
    Name,
    League,
    Nation
    Weight
FROM
    eafc26
WHERE
    League = 'Liga F Moeve'
"""
pd.read_sql(query_liga, conn)
```

```python
query_nama = """
SELECT
    Name,
    Age,
    Nation
FROM
    eafc26
WHERE
    Name Like 'Kylian%'
"""
pd.read_sql(query_nama, conn)
```
